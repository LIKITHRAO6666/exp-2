```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Simulating a week's worth of weather data (7 days)
days = 7
dates = pd.date_range("2024-12-26", periods=days, freq="D")

# Base temperature curve (sinusoidal daily temperature variation)
base_temp = 15 + 10 * np.sin(np.linspace(0, 2 * np.pi, days))

# Adding some random noise to simulate real-world variability
noise = np.random.normal(loc=0, scale=2, size=days)

# Final temperature model (base + noise)
temperature = base_temp + noise

# Create a DataFrame for the weather data
weather_data = pd.DataFrame({
    'Date': dates,
    'Temperature (°C)': temperature
})

# Plot the simulated weather data
plt.figure(figsize=(10, 6))
plt.plot(weather_data['Date'], weather_data['Temperature (°C)'], marker='o', linestyle='-', color='b', label="Temperature")
plt.fill_between(weather_data['Date'], weather_data['Temperature (°C)'], color='lightblue', alpha=0.3)
plt.title("Simulated Daily Temperature Forecast")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Print the generated weather data
print(weather_data)
```
